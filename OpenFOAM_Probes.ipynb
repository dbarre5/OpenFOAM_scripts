```python
import re

#Store text in a list of strings
with open('U') as f:
    lines = f.readlines()

#lines[0].split()

#Set some delimiters
delimiters = "(", ")"
delimiters_2 = " "
regexPattern = '|'.join(map(re.escape, delimiters))
regexPattern_2 = '|'.join(map(re.escape, delimiters_2))
tempSplit = re.split(regexPattern, lines[300])[1]
print(re.split(regexPattern_2, tempSplit))

#Create a boolean switch for checking whether you are not at the correct line in the text file
switch = False
#Create empty probe list
probeList = []
#Loops through lines in the text file
for i in range(len(lines)):
    #Check if you are at the correct line
    if switch == True:
        #Store that line in the matrix
        counter = 0
        for k in range(1,len(re.split(regexPattern, lines[i]))-2,2):
            tempSplit = re.split(regexPattern, lines[i])[k]
            if float(re.split(regexPattern_2, tempSplit)[0]) == -1e+300:
                probeList[counter].append((0.0,0.0,0.0))
            else:
                probeList[counter].append((float(re.split(regexPattern_2, tempSplit)[0]),float(re.split(regexPattern_2, tempSplit)[1]),float(re.split(regexPattern_2, tempSplit)[2])))
            counter+=1
            
    #Set the switch to True when you reach the time line
    if lines[i].split()[1] == 'Time':
        switch = True
        # Append a list inside of the list for each probe
        for j in range(2,len(lines[i-1].split())):
            probeList.append([lines[i-1].split()[j]])

            
            
probeListX = []
for j in range(len(probeList)-1):
    probeListTemp = []
    for i in range(len(probeList[0][1:])):
        probeListTemp.append(probeList[j][1:][i][0])
    probeListX.append(probeListTemp)
    
probeListY = []
for j in range(len(probeList)-1):
    probeListTemp = []
    for i in range(len(probeList[0][1:])):
        probeListTemp.append(probeList[j][1:][i][1])
    probeListY.append(probeListTemp)

probeListZ = []
for j in range(len(probeList)-1):
    probeListTemp = []
    for i in range(len(probeList[0][1:])):
        probeListTemp.append(probeList[j][1:][i][2])
    probeListZ.append(probeListTemp)
    
    
    
    
for j in range(10):
    for i in range(10*j,10*(j+1)):
        plt.plot(probeListX[i][1:],label=probeList[i][0])
    plt.legend(loc=8,bbox_to_anchor=(0.5, 0., 0.5, 0.5))
    plt.title("X_Velocity vs. Time")
    plt.show()
for i in range(100,107):
    plt.plot(probeListX[i][1:],label=probeList[i][0])
    plt.title("X_Velocity vs. Time")
    plt.legend(loc=8,bbox_to_anchor=(0.5, 0., 0.5, 0.5))
plt.show()

for j in range(10):
    for i in range(10*j,10*(j+1)):
        plt.plot(probeListY[i][1:],label=probeList[i][0])
    plt.legend(loc=8,bbox_to_anchor=(0.5, 0., 0.5, 0.5))
    plt.title("Y_Velocity vs. Time")
    plt.show()
for i in range(100,107):
    plt.plot(probeListY[i][1:],label=probeList[i][0])
    plt.title("Y_Velocity vs. Time")
    plt.legend(loc=8,bbox_to_anchor=(0.5, 0., 0.5, 0.5))
plt.show()

for j in range(10):
    for i in range(10*j,10*(j+1)):
        plt.plot(probeListZ[i][1:],label=probeList[i][0])
    plt.legend(loc=8,bbox_to_anchor=(0.5, 0., 0.5, 0.5))
    plt.title("Z_Velocity vs. Time")
    plt.show()
for i in range(100,107):
    plt.plot(probeListZ[i][1:],label=probeListZ[i][0])
    plt.title("Z_Velocity vs. Time")
    plt.legend(loc=8,bbox_to_anchor=(0.5, 0., 0.5, 0.5))
plt.show()
```
